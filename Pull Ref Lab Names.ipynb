<h1>Web Scraping in Action</h1>



```python
from bs4 import BeautifulSoup
import urllib
```

```python
html = open("Specialist and reference microbiology laboratory tests and services - GOV_UK.htm")
soup = BeautifulSoup(html, 'html.parser')

#url = r'https://www.gov.uk/guidance/specialist-and-reference-microbiology-laboratory-tests-and-services'
#response = urllib.request.urlopen(url)
#page_content = response.read()
#soup = BeautifulSoup(page_content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
tr_elements = soup.find_all('tr')
ref_names = {}
for tag in tr_elements:
    for element in tag.find_all('abbr'):
        ref_names[element.string] = element['title']
```

```python
del ref_names['ELISA']
del ref_names['PCR']
del ref_names['anti-PT']
del ref_names['CF']
del ref_names['IgG']
del ref_names['HAV']
del ref_names['HBV']
```

```python
ref_names
```

```python
from collections import defaultdict
ref_forms = defaultdict(set)
for tag in tr_elements:
    for element_abbr in tag.find_all('abbr'):
        for element_a in tag.find_all('a'):
            ref_forms[element_abbr.string].add(element_a['href'])
ref_forms
```

```python
del ref_forms['ELISA']
del ref_forms['PCR']
del ref_forms['anti-PT']
del ref_forms['CF']
del ref_forms['IgG']
del ref_forms['HAV']
del ref_forms['HBV']
```

```python
ref_forms
```

```python
import pandas as pd
```

```python
ref_forms_df = pd.Series(ref_forms, name='Ref Forms')
ref_forms_df.index.name = 'Code'
ref_forms_df.reset_index()
ref_names_df = pd.Series(ref_names, name='Ref Names')
ref_names_df.index.name = 'Code'
ref_names_df.reset_index()
ref_labs_df = pd.concat([ref_names_df,ref_forms_df], axis=1)
ref_labs_df
```

```python
ref_labs_df['Ref Forms'] = ref_labs_df['Ref Forms'].astype(str)
```

```python
ref_labs_df.to_excel('ref_labs_PHE.xlsx')
```

LETS GET THE ORGANISM NAMES! AND ASSOCIATE THEM TO REF LAB CODES!

```python
td_elements = []
for tr in tr_elements:
    td_elements.append(tr.find_all('td'))
```

```python
org_code = {}
for row in td_elements:
    if len(row) > 0:
        code = row[len(row)-1].find_all("abbr")
        if len(code) != 0:
            code = code[0].string
            org_code[row[0].string] = code
org_code
```

```python
org_codes = pd.Series(org_code, name='Code')
org_codes.index.name = 'Organism'
org_codes.reset_index()
```

```python
org_codes.to_excel('org_ref_lab_codes.xlsx')
```
